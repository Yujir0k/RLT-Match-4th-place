# ML package overview

Этот ноутбук фиксирует состав ML-части проекта `Hybrid Lot Matcher`.

Модель работает локально и не использует внешние API. На вход принимает номенклатурную матрицу поставщика, а на выходе возвращает top-3 релевантных `pn_lot` для каждого товара.


## README из переданного ML-пакета


# Hybrid Lot Matcher Package

Это передаваемый пакет для интеграции в сайт.

Что внутри:

- готовый гибридный matcher `TF-IDF + char n-grams + dense embeddings + rules`
- готовые артефакты индекса и эмбеддингов
- runtime bundle с уже подготовленной базой закупок
- простой Python API и CLI для пакетного прогноза

## Основная идея

На вход подается CSV с товарами покупателя.
На выходе для каждого товара возвращаются 3 лучших лота в приоритетном порядке.

Система уже заранее знает базу закупок.
Во время online-запроса она не переобучается и не перестраивает индекс.

## Основные файлы

- `hybrid_matcher_runtime.py` — основной runtime-класс `HybridLotMatcher`
- `predict_batch.py` — CLI для батчевого запуска
- `ml_baseline.py` — исходная логика, на которой собран runtime
- `config/model_config.json` — конфиг скоринга и пути к артефактам
- `artifacts/index/` — TF-IDF, char n-grams, SVD и lot embeddings
- `artifacts/runtime/runtime_bundle.pkl` — готовый runtime bundle

## Быстрый запуск

```bash
cd handoff/hybrid_lot_matcher_package
python3 -m pip install -r requirements.txt
python3 predict_batch.py \
  --input-csv /path/to/buyer_input.csv \
  --output-csv /path/to/buyer_output.csv \
  --top-lots 3
```

## Требования к входному CSV

Обязательные колонки:

- `id`
- `Категория`
- `Наименование`
- `Характеристики`

Поддерживаются разделители `;` и `,`.
Также поддерживается опечатка `Наименвоание`.

## Формат результата

Основные поля на выходе:

- `seller_id`
- `seller_category`
- `seller_name`
- `lot_rank`
- `pn_lot`
- `lot_subject`
- `matched_unit_name`
- `confidence`
- `score_100`
- `matched_terms`
- `explanation_short`

## Для разработчика сайта

Минимальный сценарий интеграции:

```python
from pathlib import Path
from hybrid_matcher_runtime import HybridLotMatcher

matcher = HybridLotMatcher(Path("/absolute/path/to/handoff/hybrid_lot_matcher_package"))
result = matcher.predict("/absolute/path/to/buyer_input.csv", top_lots=3)
```

`result` — это `pandas.DataFrame`, который уже можно отдавать дальше в API или сохранять в БД.
